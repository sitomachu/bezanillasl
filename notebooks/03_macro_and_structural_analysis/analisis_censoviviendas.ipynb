# Análisis del censo de viviendas 2021

Objetivo principal:
- Calcular el tamaño medio según el tipo de vivienda en Santa Cruz de Bezana

Carpeta de datos esperada:
- `data/raw/INE/CensoViviendas_2021.csv`


## 1. Setup: imports, rutas y configuración

Ajusta los nombres de archivo si has usado otra convención.

```python
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 140)

DATA_DIR = Path("../../data/raw/INE")

CENSO_VIVIENDAS = DATA_DIR / "CensoViviendas_2021.csv"
```

## 2. Carga de datos

```python
df = pd.read_csv(CENSO_VIVIENDAS,
                 sep=',',
                 index_col = 0)

print(f"Dataset con {df.shape[0]} registros y {df.shape[1]} columnas\n")
df.head()
```

## 3. Procesado

### Renombrar columnas y describir con datos del INE

```python
df.index.name = "id"

column_map = {
                                            # Códigos territoriales INE
    "CPRO": "codigo_provincia",             # Código de provincia (INE). Cantabria = 39
    "CMUN": "codigo_municipio",             # Código de municipio (INE). Santa Cruz de Bezana = 073
    "TIPO_VIV": "tipo_vivienda",            # Tipo de vivienda: 1 = vivienda principal, 0 = no principal
    "TENEN_VIV": "regimen_tenencia",        # 2 = propiedad, 3 = alquiler, 4 = otras formas
    "RT_imp": "imputacion_RT",              # Indicador de imputación de renta: 0 = no imputado, 1 = imputado
    "SUPERF": "superficie_util",            # Superficie útil en m². Valor 1000 indica truncamiento (>= 1000 m²)
    "PLANTAS_sr": "plantas_sobre_rasante",  # Número de plantas sobre rasante. 10 indica top-code (>= 10 plantas)
    "PLANTAS_br": "plantas_bajo_rasante",   # Número de plantas bajo rasante. 2 indica top-code (>= 2 plantas)
    "TIPO_EDIF": "tipo_edificio",           # Tipo de edificio:
                                                # 1 = edificio residencial con 1 vivienda (unifamiliar)
                                                # 2 = edificio residencial con 2 viviendas (bifamiliar)
                                                # 3 = edificio residencial con ≥3 viviendas (plurifamiliar)
                                                # 4 = edificio no residencial con viviendas
    "ANOCO": "ano_construccion",            # Año de construcción. 1900 indica truncamiento (<= 1900)
    "TAM_HOG_5": "tamano_hogar"             # Número de personas en el hogar. 5 indica top-code (>= 5 personas)
}

df_rename = df.rename(columns=column_map).copy()

df_rename.head(10)
```

Aislamiento de dataset para Cantabria

```python
df_cantabria = df_rename.loc[df_rename["codigo_provincia"] == 39].copy()
print(f"Dataset de Cantabria con {df_cantabria.shape[0]} registros y {df_cantabria.shape[1]} columnas\n")
df_cantabria.head()
```

Aislamiento de dataset para Bezana

```python
df_bezana = df_rename.loc[(df_rename["codigo_provincia"] == 39) & (df_rename["codigo_municipio"] == 73)].copy()
print(f"Dataset de Bezana con {df_bezana.shape[0]} registros y {df_bezana.shape[1]} columnas\n")
df_bezana.head()
```

Función de filtrado (para ambos)

```python
def filtrar_base(df_in):
    d = df_in.copy()
    d = d[d["tipo_vivienda"] == 1]          # viviendas principales
    d = d[d["tipo_edificio"].isin([1,2,3])] # solo viviendas residenciales
    d = d[d["superficie_util"].notna()]     # excluir sin superficie
    d = d[d["superficie_util"] != 1000]     # excluir top code
    return d
```

```python
df_cantabria_clean = filtrar_base(df_cantabria)
df_bezana_clean = filtrar_base(df_bezana)
```

### Agrupaciones relevantes

```python
map_tip = {
    1: "unifamiliar_1_vivienda",
    2: "bifamiliar_2_viviendas",
    3: "plurifamiliar_bloque"
}

for d in (df_cantabria_clean, df_bezana_clean):
    d["tipologia_edificio"] = d["tipo_edificio"].map(map_tip)

def resumen_superficie(df_in):
    return (df_in
            .groupby("tipologia_edificio")["superficie_util"]
            .agg(n="count", media="mean", mediana="median")
            .reset_index()
           )

res_cantabria = resumen_superficie(df_cantabria_clean)
res_bezana    = resumen_superficie(df_bezana_clean)

print(f"Resumen del dataset de Cantabria:\n{res_cantabria}")
print(f"\nResumen del dataset de Bezana:\n{res_bezana}")
```
