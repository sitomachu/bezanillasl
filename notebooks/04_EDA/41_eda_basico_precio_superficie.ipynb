# EDA Basico: Sale y Rent

Analisis exploratorio corto y separado para `sale_homes_clean.csv` y `rent_homes_clean.csv`.

```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd

plt.style.use("seaborn-v0_8-whitegrid")

def find_project_root(start_path: Path) -> Path:
    for candidate in [start_path, *start_path.parents]:
        if (candidate / "data" / "processed" / "idealistaAPI").exists():
            return candidate
    raise FileNotFoundError("No se encontro la raiz del proyecto con data/processed/idealistaAPI")

PROJECT_ROOT = find_project_root(Path.cwd().resolve())
DATA_DIR = PROJECT_ROOT / "data" / "processed" / "idealistaAPI"

SALE_PATH = DATA_DIR / "sale_homes_clean.csv"
RENT_PATH = DATA_DIR / "rent_homes_clean.csv"

sale_df = pd.read_csv(SALE_PATH)
rent_df = pd.read_csv(RENT_PATH)

print(f"SALE shape: {sale_df.shape}")
print(f"RENT shape: {rent_df.shape}")
```

```python
columnas = [
    "precio",
    "superficie_construida_m2",
    "numero_dormitorios",
    "numero_banos",
]

def mostrar_resumen(nombre: str, df: pd.DataFrame) -> None:
    resumen = (
        df[columnas]
        .describe()
        .T[["count", "mean", "std", "min", "25%", "50%", "75%", "max"]]
        .round(2)
    )
    nulos = df[columnas].isna().sum().to_frame("nulos")

    print(f"\n=== {nombre} ===")
    print("Resumen estadistico:")
    print(resumen.to_string())
    print("Nulos por variable:")
    print(nulos.to_string())

mostrar_resumen("SALE", sale_df)
mostrar_resumen("RENT", rent_df)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].scatter(sale_df["superficie_construida_m2"], sale_df["precio"], alpha=0.35, s=18)
axes[0].set_title("SALE: Precio vs Superficie")
axes[0].set_xlabel("Superficie construida (m2)")
axes[0].set_ylabel("Precio")

axes[1].scatter(rent_df["superficie_construida_m2"], rent_df["precio"], alpha=0.35, s=18, color="darkorange")
axes[1].set_title("RENT: Precio vs Superficie")
axes[1].set_xlabel("Superficie construida (m2)")
axes[1].set_ylabel("Precio")

plt.tight_layout()
plt.show()
```
